```python
import pandas as pd
import numpy as np
import plotly.express as px
import sys
sys.path.append('..')
from utils.text_cleaning import clean_text

# Load your raw dataset here
df = pd.read_csv('../data/raw/YOUR_DATASET.csv')
df.head()
```

```python
print(f'Shape: {df.shape}')
print(f'Columns: {df.columns.tolist()}')
df.info()
```

```python
# Null check
df.isnull().sum()
```

```python
# Duplicate check
print(f'Duplicates: {df.duplicated().sum()}')
```

```python
# Text length distribution
df['text_length'] = df['text'].astype(str).apply(len)
fig = px.histogram(df, x='text_length', nbins=50, title='Text Length Distribution')
fig.show()
```

```python
# If labels exist — class distribution
if 'label' in df.columns or 'sentiment' in df.columns:
    label_col = 'label' if 'label' in df.columns else 'sentiment'
    fig = px.bar(df[label_col].value_counts().reset_index(),
                 x=label_col, y='count', title='Label Distribution')
    fig.show()
```
